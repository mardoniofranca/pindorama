```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define a função
def f(x):
    return (x - 8) * (x + 8) * (x ** 2)
```

```python
# Cria um array de valores de x (de -5 a 5, com 400 pontos)
x = np.linspace(-10, 10, 400)
# Calcula os valores de y correspondentes
y = f(x)
```

```python
# Configura o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x) = (x-8)(x+8)(x^2)$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)  # Linha do eixo x
plt.axvline(0, color='black', linewidth=0.5)  # Linha do eixo y
plt.title('Gráfico de $f(x) = (x-8)(x+8)(x^2)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Função e sua derivada
#x^4 - 64*x^2
def f(x):
    return (x - 8) * (x + 8) * (x ** 2)
#4x^3 - 128*x^2
def df(x):  # Derivada de f(x)
    return 4 * x**3 - 128 * x
```

```python
# Parâmetros do Gradiente Descendente
alpha = 0.0001  # Taxa de aprendizagem
max_iter = 1000  # Número máximo de iterações
tol = 1e-6  # Tolerância para parada (convergência)
x = 10  # Ponto inicial (pode ser qualquer valor)

# Armazenamento para visualização
x_history = [x]
f_history = [f(x)]

# Loop do Gradiente Descendente
for _ in range(max_iter):
    x_new = x - alpha * df(x)  # Atualização de x
    x_history.append(x_new)
    f_history.append(f(x_new))
    
    # Verifica convergência
    if abs(x_new - x) < tol:
        break
    x = x_new

# Resultado
print(f"Ponto de mínimo encontrado: x = {x:.4f}")
print(f"Valor da função no mínimo: f(x) = {f(x):.4f}")

# Plot da trajetória do gradiente descendente
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = (x-8)(x+8)(x^2)$', color='blue')
plt.scatter(x_history, f_history, color='red', s=50, label='Trajetória do Gradiente')
plt.title('Gradiente Descendente para Encontrar o Mínimo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
