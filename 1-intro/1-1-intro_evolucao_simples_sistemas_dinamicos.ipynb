```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mapa_logistico(A, x0, n):
    valores = [x0]      
    for _ in range(n - 1): 
        x_next = A * valores[-1] * (1 - valores[-1]) 
        valores.append(x_next)      
    return valores
```

```python
# Parâmetros
A = 2.8; x0_1 = 0.8; n = 100; x0_2 = 0.1

# Gerando os valores
valores_x1 = mapa_logistico(A, x0_1, n)
valores_x2 = mapa_logistico(A, x0_2, n)

# Ajustando o tamanho da figura (mais comprido)
plt.figure(figsize=(12, 5))  # Aumenta a largura do gráfico

# Criando o gráfico
plt.plot(range(n), valores_x1, marker='o', linestyle='-', color='blue', markersize=2, linewidth=0.8, label=f"x0 = {x0_1}")
plt.plot(range(n), valores_x2, marker='o', linestyle='-', color='red', markersize=2, linewidth=0.8, label=f"x0 = {x0_2}")

# Personalizando o gráfico
plt.xlabel("t (tempo)")
plt.ylabel("x(t)")
plt.title(f"Mapa Logístico para A={A}")
plt.legend()
plt.grid()
plt.show()
```
